```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Adidas = pd.read_excel('Adidas US Sales Datasets.xlsx')
```

```python
Adidas
```

```python
# Drop rows 1 to 3
Adidas = Adidas.drop(Adidas.index[0:3])

# Drop the first column
Adidas = Adidas.iloc[:, 1:]

# Reset the first row as the header
new_header = Adidas.iloc[0]
Adidas = Adidas[1:]
Adidas.columns = new_header

# Reset the index if needed
Adidas = Adidas.reset_index(drop=True)

# Print the modified DataFrame
Adidas
```

```python
Adidas.info()
```

```python

# drop duplicate data
Adidas=Adidas.drop_duplicates()

# convert data type
Adidas['Invoice Date']=pd.to_datetime(Adidas['Invoice Date'])

# Convert specific columns to float
cols_to_convert = ['Price per Unit', 'Units Sold', 'Total Sales', 'Operating Profit', 'Operating Margin']
Adidas[cols_to_convert] = Adidas[cols_to_convert].astype(float)
Adidas.info()
```

```python
#describe data in statistics
Adidas.describe()
```

```python
Adidas['Total Sales'] = Adidas['Total Sales']/10
Adidas['Operating Profit'] = Adidas['Operating Profit']/10

Adidas.head()
```

```python
Adidas.to_csv('cleaned_data.csv', index=False)
```
